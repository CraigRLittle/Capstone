```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt

# Import Files
ed19 = pd.read_csv('./2019Education.csv')
ed22 = pd.read_csv('./2022Education.csv')
el20 = pd.read_csv('./2020Election.csv')
```

```python
# Election dataframe manipulation for graph

# Remove ',' from numbers in columns
el20['TOTAL'].replace(',','',regex=True,inplace=True)
el20['TOTAL'] = pd.to_numeric(el20['TOTAL'])

el20['BIDEN'].replace(',','',regex=True,inplace=True)
el20['BIDEN'] = pd.to_numeric(el20['BIDEN'])

el20['TRUMP'].replace(',','',regex=True,inplace=True)
el20['TRUMP'] = pd.to_numeric(el20['TRUMP'])

el20['OTHER'].replace(',','',regex=True,inplace=True)
el20['OTHER'] = pd.to_numeric(el20['OTHER'])

el20['BIDEN %'].replace('%','',regex=True,inplace=True)
el20['BIDEN %'] = pd.to_numeric(el20['BIDEN %'])

el20['TRUMP %'].replace('%','',regex=True,inplace=True)
el20['TRUMP %'] = pd.to_numeric(el20['TRUMP %'])
```

```python
# Add education plots for '19 Math %

# Sort list
ed19.sort_values('ORDER', ascending=True, inplace=True)

#Assign plots for '19 Math %
a = ed19['MATH %']
```

```python
# Add education plots for '19 Math %

# Sort list
ed19.sort_values('ORDER', ascending=True, inplace=True)

#Assign plots for '19 Math %
b = ed19['READING %']
```

```python
# Add education plots for '22 Math %

# Sort list
ed22.sort_values('ORDER', ascending=True, inplace=True)

#Assign plots for '19 Math %
c = ed22['MATH %']
```

```python
# Add education plots for '19 Math %

# Sort list
ed22.sort_values('ORDER', ascending=True, inplace=True)

#Assign plots for '19 Math %
d = ed22['READING %']
```

```python
# Add plots for Election and format plat graph

# Sort list
el20.sort_values('BIDEN %', ascending=True, inplace=True)

# Assign plots for vote %
v = el20['BIDEN %']
x = el20['STATE / 2020']
y = el20['TRUMP %']

# Label and format graph
plt.figure(figsize=(14, 7.5))
plt.xlabel('States (Red) Republican (Blue) Democrat')
plt.ylabel('Vote % (excluding 3rd party votes)')
plt.grid('minor')
plt.xticks(rotation='vertical')
plt.locator_params(axis='y', nbins=40)
plt.title("Republican / Democrat Vote % per State\n'19-'22 Math and Reading % / 500\n 4th Grade")

# Plot line
plt.plot(x, a, 'g-', label = "'19 MATH %")
plt.plot(x, c, 'y-', label = "'22 MATH %")
plt.plot(x, b, 'k-', label = "'19 READING %")
plt.plot(x, d, 'm-', label = "'22 READING %")
plt.plot(x, v, 'b--', label = 'DEMOCRAT VOTE %')
plt.plot(x, y, 'r--', label = 'REPUPLICAN VOTE %')

# Sort for Replublican vote
el20.sort_values('TRUMP %', ascending=False, inplace=True)

# Add and locate Legend
plt.legend(loc = 'upper center')

# Get the x tick labels
xtick_labels = plt.xticks()[1]

# Set the color for each tick label
for i, label in enumerate(xtick_labels):
    if i < len(xtick_labels) / 2:
        plt.setp(label, color='red')
    else:
        plt.setp(label, color='blue')
```
