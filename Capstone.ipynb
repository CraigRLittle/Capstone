```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt

# Import Files
ed19 = pd.read_csv('./2019Education.csv')
ed22 = pd.read_csv('./2022Education.csv')
el20 = pd.read_csv('./2020Election.csv')
```

```python
# Election dataframe manipulation for graph

# Remove ',' from numbers in columns
el20['TOTAL'].replace(',','',regex=True,inplace=True)
el20['TOTAL'] = pd.to_numeric(el20['TOTAL'])

el20['BIDEN'].replace(',','',regex=True,inplace=True)
el20['BIDEN'] = pd.to_numeric(el20['BIDEN'])

el20['TRUMP'].replace(',','',regex=True,inplace=True)
el20['TRUMP'] = pd.to_numeric(el20['TRUMP'])

el20['OTHER'].replace(',','',regex=True,inplace=True)
el20['OTHER'] = pd.to_numeric(el20['OTHER'])

el20['BIDEN %'].replace('%','',regex=True,inplace=True)
el20['BIDEN %'] = pd.to_numeric(el20['BIDEN %'])

el20['TRUMP %'].replace('%','',regex=True,inplace=True)
el20['TRUMP %'] = pd.to_numeric(el20['TRUMP %'])

# Sort by Biden Votes %
el20.sort_values('BIDEN %', ascending=False, inplace=True)
```

```python
# Add education plots for '19 Math %

# Sort list
ed19.sort_values('MATH %', ascending=True, inplace=True)

#Assign plots for '19 Math %
a = ed19['MATH %'][28::]
a2 = ed19['MATH %'][:28:]
```

```python
# Add education plots for '19 Math %

# Sort list
ed19.sort_values('READING %', ascending=True, inplace=True)

#Assign plots for '19 Math %
b = ed19['READING %'][28::]
b2 = ed19['READING %'][:28:]
```

```python
# Add education plots for '22 Math %

# Sort list
ed22.sort_values('MATH %', ascending=True, inplace=True)

#Assign plots for '19 Math %
c = ed22['MATH %'][28::]
c2 = ed22['MATH %'][:28:]
```

```python
# Add education plots for '19 Math %

# Sort list
ed22.sort_values('READING %', ascending=True, inplace=True)

#Assign plots for '19 Math %
d = ed22['READING %'][28::]
d2 = ed22['READING %'][:28:]
```

```python
# Add plots for Election and format plat graph

# Sort list
el20.sort_values('BIDEN %', ascending=True, inplace=True)

# Assign plots for vote %
x = el20['STATE / 2020'][28::]
x2 = el20['STATE / 2020'][:28:]
y = el20['BIDEN %'][28::]
y2 = el20['TRUMP %'][:28:]

# Label and format graph
plt.figure(figsize=(14, 7.5))
plt.xlabel('States')
plt.ylabel('Vote %')
plt.grid('minor')
plt.xticks(rotation='vertical')
plt.locator_params(axis='y', nbins=40)
plt.title("Republican / Democrat Vote % per State\n'19 Math and Reading % / 500")

# Plot line
plt.plot(x2, y2, 'r-', label = 'REPUPLICAN VOTE %')
plt.plot(x, y, 'b-', label = 'DEMOCRAT VOTE %')
plt.plot(x, a, 'g-', label = "'19 MATH %")
plt.plot(x2, a2, 'g-')
plt.plot(x, b, 'k-', label = "'19 READING %")
plt.plot(x2, b2, 'k-')
plt.plot(x, c, 'y-', label = "'22 MATH%")
plt.plot(x2, c2, 'y-')
plt.plot(x, d, 'm-', label = "'22 READING %")
plt.plot(x2, d2, 'm-')

# Add and locate Legend
plt.legend(loc = 'upper center')
```
