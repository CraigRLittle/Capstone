```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt

# Import files
income19 = pd.read_csv('./2019Income.csv')
income22 = pd.read_csv('./2022Income.csv')
el20 = pd.read_csv('2020Election.csv')
```

```python
el20
```

```python
# Election dataframe manipulation for graph

# Remove ',' from numbers in columns
el20['TOTAL'].replace(',','',regex=True,inplace=True)
el20['TOTAL'] = pd.to_numeric(el20['TOTAL'])

el20['BIDEN'].replace(',','',regex=True,inplace=True)
el20['BIDEN'] = pd.to_numeric(el20['BIDEN'])

el20['TRUMP'].replace(',','',regex=True,inplace=True)
el20['TRUMP'] = pd.to_numeric(el20['TRUMP'])

el20['OTHER'].replace(',','',regex=True,inplace=True)
el20['OTHER'] = pd.to_numeric(el20['OTHER'])

el20['BIDEN %'].replace('%','',regex=True,inplace=True)
el20['BIDEN %'] = pd.to_numeric(el20['BIDEN %'])

el20['TRUMP %'].replace('%','',regex=True,inplace=True)
el20['TRUMP %'] = pd.to_numeric(el20['TRUMP %'])
```

```python
# Add plots for Election and format plat graph

# Sort list
el20.sort_values('BIDEN %', ascending=True, inplace=True)

# Assign plots for vote %
v = el20['BIDEN %']
x = el20['STATE / 2020']
y = el20['BIDEN %']

# Label and format graph
plt.figure(figsize=(14, 7.5))
plt.xlabel('States')
plt.ylabel('Income ($)')
plt.grid('minor')
plt.xticks(rotation='vertical')
plt.locator_params(axis='y', nbins=40)
plt.title("Republican / Democrat Vote % per State\nMedian Income 2019 vs 2022")

# Plot line
# plt.plot(x, c, 'y-', label = "'22 MATH %")
# plt.plot(x, d, 'm-', label = "'22 READING %")
# plt.plot(x, a, 'g-', label = "'19 MATH %")
# plt.plot(x, b, 'k-', label = "'19 READING %")
plt.plot(x, v, 'b--', label = 'DEMOCRAT VOTE %')

# Sort for Replublican vote
el20.sort_values('TRUMP %', ascending=False, inplace=True)

# Assign plot for Republican vote
y = el20['TRUMP %']

# Plot Republican votes
plt.plot(x, y, 'r--', label = 'REPUPLICAN VOTE %')

# Add and locate Legend
plt.legend(loc = 'upper center')
```
